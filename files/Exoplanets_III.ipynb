```python
# Import some python basic  modules
import numpy as np
import matplotlib.pyplot as plt

# Import the module that has the transit model (forTrans)
from PyAstronomy.modelSuite import forTrans as ft
```

```python
def plotting(per, a, p , T0, color):
    global Rsun
    Rsun = 696340e3
    planet =  ft.MandelAgolLC(orbit="circular", ld="quad")

    planet["per"] = per

    planet["i"] = 90

    planet["a"] = a

    planet["T0"] = T0

    # planet["p"] = Rp /RSun
    planet["p"] = p

    planet["linLimb"] = 0.5
    planet["quadLimb"] = 0.2
    planet["b"] = 0.

    time = np.linspace(0, 20, 1000)

    stellar_flux = planet.evaluate(time)

    plt.title ("2-star Transit curve", fontsize = 15)
    plt.xlabel("Time (days)", fontsize = 15)
    plt.ylabel("Stellar relative flux", fontsize = 15)

    # put grid on the graph
    plt.grid(color='grey', linestyle=':', linewidth=0.2)
    plt.grid(True)

    # Let's see what happened ...
    plt.plot(time, stellar_flux, color= color, linewidth=1)

# ### System I

# Planet 1:per1=6 days, a1=2, p1= 0.2, T0=3

# Planet 2:per1=3 days, a1=3, p1= 0.1, T0=3.5

# ### System II

# Planet 1:per1=6 days, a1=2, p1= 0.2, T0=3

# Planet 2:per1=1.5 days, a1=3, p1= 0.3, T0=3.5

planet1 = plotting(6, 2, 0.2, 3, 'blue')

planet2 = plotting(3, 3, 0.1, 3.5, 'red')

plt.show()
```

```python
def plotting(per1, a1, p1 , T01, per2, a2, p2, T02):
    global Rsun
    Rsun = 696340e3
    planet1 =  ft.MandelAgolLC(orbit="circular", ld="quad")

    planet1["per"] = per1

    planet1["i"] = 90

    planet1["a"] = a1

    planet1["T0"] = T01

    # planet["p"] = Rp /RSun
    planet1["p"] = p1

    planet1["linLimb"] = 0.5
    planet1["quadLimb"] = 0.2
    planet1["b"] = 0.

    time = np.linspace(0, 8, 1000)

    stellar_flux1 = planet1.evaluate(time)

    planet2 =  ft.MandelAgolLC(orbit="circular", ld="quad")

    planet2["per"] = per2

    planet2["i"] = 90

    planet2["a"] = a2

    planet2["T0"] = T02

    # planet["p"] = Rp /RSun
    planet2["p"] = p2

    planet2["linLimb"] = 0.5
    planet2["quadLimb"] = 0.2
    planet2["b"] = 0.

    time = np.linspace(0, 8, 1000)

    stellar_flux2 = planet2.evaluate(time)

    plt.title ("2-star Transit curve", fontsize = 15)
    plt.xlabel("Time (days)", fontsize = 15)
    plt.ylabel("Stellar relative flux", fontsize = 15)

    # put grid on the graph
    plt.grid(color='grey', linestyle=':', linewidth=0.2)
    plt.grid(True)

    # Let's see what happened ...
    plt.plot(time, stellar_flux1 + stellar_flux2 - 1, color= 'black', linewidth=1)
    plt.plot(time, stellar_flux1, color= 'blue', linewidth=1, linestyle='dotted')
    plt.plot(time, stellar_flux2, color= 'red', linewidth=1, linestyle='dotted')

# ### System I

# Planet 1:per1=6 days, a1=2, p1= 0.2, T0=3

# Planet 2:per1=3 days, a1=3, p1= 0.1, T0=3.5

# ### System II

# Planet 1:per1=6 days, a1=2, p1= 0.2, T0=3

# Planet 2:per1=1.5 days, a1=3, p1= 0.3, T0=3.5

planet = plotting(6, 2, 0.2, 3, 3, 3, 0.1, 3.5)


plt.show()
```
